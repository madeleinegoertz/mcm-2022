```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("C:\\Users\\madie\\mcm-2022\\tokyo.csv")
df = df.rename(columns={"x":"longitude","y":"latitude","ele":"elevation"})
df = df.drop(columns={"track_fid","track_seg_id"})
# we computed that the latitude distortion factor is 111111 and the longitude distortion factor is 90795
df.head()
```

```python
plt.scatter(x=df["track_seg_point_id"],y=df["elevation"])
```

```python
min_long = np.min(df["longitude"])
max_long = np.max(df["longitude"])
min_lat = np.min(df["latitude"])
max_lat = np.max(df["latitude"])

# spherical coordinate conversion
# Varying latitude

def get_distortion_factor(v_0, v_1, spread):
    r = 6378100 #radius of earth in meters
    sph_0 = (r,(90-v_0[1])*math.pi/180, (360-v_0[0])*math.pi/180)
    sph_1 = (r,(90-v_1[1])*math.pi/180, (360-v_1[0])*math.pi/180)
    rect_0 = (sph_0[0]*math.sin(sph_0[1])*math.cos(sph_0[2]),
              sph_0[0]*math.sin(sph_0[1])*math.sin(sph_0[2]),
              sph_0[0]*math.cos(sph_0[1]))
    rect_1 = (sph_1[0]*math.sin(sph_1[1])*math.cos(sph_1[2]),
              sph_1[0]*math.sin(sph_1[1])*math.sin(sph_1[2]),
              sph_1[0]*math.cos(sph_1[1]))
    # angle between the two vectors
    alpha = math.acos((rect_0[0]*rect_1[0]+rect_0[1]*rect_1[1]+rect_0[2]*rect_1[2])/(r*r))
    # arclength along great circle
    s = r*alpha
    distortion_factor = s / spread
    return distortion_factor
```

```python
# latitude
v_0 = (min_long, min_lat)
v_1 = (min_long, max_lat)
dist_lat = get_distortion_factor(v_0,v_1,abs(v_0[1]-v_1[1]))
v_2 = (min_long, min_lat)
v_3 = (max_long, min_lat)
dist_long = get_distortion_factor(v_2,v_3,abs(v_2[0]-v_3[0]))
```

```python
df["lat_adj"] = (df["latitude"] - min_lat)*dist_lat
df["long_adj"] = (df["longitude"] - min_long)*dist_long
df["long_diff"] = df["long_adj"].diff()
df["lat_diff"] = df["lat_adj"].diff()
df["dist_btwn"] = np.sqrt(df["long_diff"]*df["long_diff"]+df["lat_diff"]*df["lat_diff"])
df["cum_dist"] = df["dist_btwn"].cumsum()
df["theta"] = np.arctan(df["elevation"].diff()/df["cum_dist"].diff())
df["theta_avg_10"] = df["theta"].rolling(10).mean()
```
